Popularity Based Recommender System
top 50 books based on highest avg rating but with mini 250 no. of ratings

```python
ratings_with_names = ratings.merge(books, on="ISBN")
ratings_with_names
```

```python
ratings_with_names
```

```python
num_rating_df = ratings_with_names.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
# avg_rating_df = ratings_with_names.groupby('Book-Title').mean()['Book-Rating'].reset_index()
# avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
# avg_rating_df
```

```python
# Convert 'Book-Rating' column to numeric explicitly
ratings_with_names['Book-Rating'] = pd.to_numeric(ratings_with_names['Book-Rating'], errors='coerce')

# Remove rows where 'Book-Rating' is NaN after conversion
ratings_with_names = ratings_with_names.dropna(subset=['Book-Rating'])

# Calculate average rating
avg_rating_df = ratings_with_names.groupby('Book-Title')['Book-Rating'].mean().reset_index()

# Rename the columns
avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)

# Print the resulting DataFrame
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on="Book-Title")
popular_df
```

```python
popular_df = popular_df[popular_df["num_ratings"]>=250].sort_values("avg_rating",ascending=False).head(50)
popular_df
```

```python
popular_df = popular_df.merge(books,on="Book-Title")
```

```python
popular_df.drop_duplicates("Book-Title")
```

```python
popular_df = popular_df[["Book-Title","Book-Author","Image-URL-M","num_ratings","avg_rating"]]
popular_df
```

Collabarative Filtering Based Recommender System
using books with min 50 ratings and users who have given 200 ratings to different books

```python
# user and how many books he has rated
x = ratings_with_names.groupby("User-ID").count()["Book-Rating"] > 200
good_users = x[x].index
```

```python
good_users
```

```python
filtered_rating = ratings_with_names[ratings_with_names["User-ID"].isin(good_users)]
filtered_rating
```

```python
y = filtered_rating.groupby("Book-Title").count()["Book-Rating"] >= 50
famous_books = y[y].index
```

```python
famous_books
```

```python
with open("book_title.pkl", "wb") as f:
  pickle.dump(famous_books,f)
```

```python
image_url = filtered_rating[filtered_rating["Book-Title"].isin(famous_books)]
```

```python
image_url = image_url.sort_values('Book-Title').drop_duplicates('Book-Title')
```

```python
image_url
```

```python
url = image_url[['Book-Title','Image-URL-L', 'Image-URL-S', 'Image-URL-M']]
url
```

```python
with open("image_url.pkl", "wb") as f:
  pickle.dump(url,f)
```

```python
final_ratings = filtered_rating[filtered_rating["Book-Title"].isin(famous_books)]
final_ratings
```

```python
with open("final_ratings.pkl", "wb") as f:
  pickle.dump(final_ratings,f)
```

```python
pt = final_ratings.pivot_table(index="Book-Title",columns="User-ID",values="Book-Rating")
```

```python
pt.fillna(0,inplace=True)
pt
```

```python
import pickle
with open("book_pivot.pkl", "wb") as f:
  pickle.dump(pt,f)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score  = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x:x[1], reverse=True)[1:11] # enumerate shows index along with matrix values
    
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('Message in a Bottle')
```
